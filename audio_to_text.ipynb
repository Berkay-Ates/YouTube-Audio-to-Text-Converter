```python
from pytube import YouTube
from pydub import AudioSegment
import speech_recognition as sr
from pathlib import Path 
import os 
```

```python
def convert_audio_to_text(wav_dosya_yolu):
    r = sr.Recognizer()

    with sr.AudioFile(wav_dosya_yolu) as source:
        audio_data = r.record(source)

        try:
            metin = r.recognize_google(audio_data, language="tr-TR")  # Türkçe için dil kodu
            return metin
        except sr.UnknownValueError:
            print("Ses algilanamadi.")
        except sr.RequestError as e:
            print(f"Hata olustu: {e}")
```

```python
def download_video_audio(url):
    # Download the video using pytube
    youtube = YouTube(url)
    video_stream = youtube.streams.filter(only_audio=True).first()
    mp3_file_name = f'{video_stream.default_filename.split(".")[0]}.mp3'
    video_stream.download(filename=mp3_file_name)

    try:
        sound = AudioSegment.from_file(mp3_file_name)
        sound.export(f'{mp3_file_name.split(".")[0]}.wav', format="wav")
        txt = convert_audio_to_text(f'{mp3_file_name.split(".")[0]}.wav')
        
        if txt:
            with open("sonuc.txt", "w", encoding="utf-8") as dosya:
                dosya.write(txt)
                print("Metin dosyasi oluşturuldu: sonuc.txt")
        else:
            print("Metin çevrimi yapilamadi.")
    except Exception as e:
        print(str(e))
```

```python
def fromm4a_to_wav(path:str):
    try:
        sound = AudioSegment.from_file(path,format='m4a')
        sound.export(f'{path.split(".")[0]}.wav',format='wav')
        
    except Exception as e:
        print(str(e))
```

```python
def sesi_metne_cevir(wav_dosya_yolu):
    r = sr.Recognizer()

    with sr.AudioFile(wav_dosya_yolu) as source:
        audio_data = r.record(source)

        try:
            metin = r.recognize_google(audio_data, language="tr-TR")  # Türkçe için dil kodu
            return metin
        except sr.UnknownValueError:
            print("Ses algılanamadı.")
        except sr.RequestError as e:
            print(f"Hata oluştu: {e}")
```

```python
def buyuk_wav_dosyasini_isle(wav_dosya_yolu, parca_sure_saniye=10):
    ses_dosyasi = AudioSegment.from_wav(wav_dosya_yolu)
    parca_sayisi = len(ses_dosyasi) // (parca_sure_saniye * 1000) + 1

    for i in range(parca_sayisi):
        baslangic_zaman = i * parca_sure_saniye * 1000
        bitis_zaman = (i + 1) * parca_sure_saniye * 1000

        parca = ses_dosyasi[baslangic_zaman:bitis_zaman]

        parca.export("temp_parca.wav".format(i), format="wav")
        metin = sesi_metne_cevir("temp_parca.wav")

        if metin:
            with open(f"{wav_dosya_yolu.split('.')[0]}.txt", "a", encoding="utf-8") as dosya:
                dosya.write(metin+"\n")
            print("Parça {} işlendi ve metin dosyası oluşturuldu".format(i))
        else:
            print("Parça {} işlenirken bir hata oluştu.".format(i))
```

```python
buyuk_wav_dosyasini_isle('Katarsis - Barış Atay “O Yaşta Gözaltına Alınmak Beni Şaşırtmadı Gururum Kırıldı”.wav', parca_sure_saniye=120)
```
